```python
# 18/11/2k18 ------------------------------------------------------------------------------
#Programador: Carlos Prado-----------------------------------------------------------------
# 01-Contenido: Repaso gerenal de librerias para graficar en python
# Funciones Trigonométricas sen cos tang
#librería visual - https://matplotlib.org
import matplotlib.pyplot as plt
#-------------------------------------------------
#Librería matemática
import numpy as np
import math 


```

<b>Fx(Sen(x))</b>

```python
# Definimos el array dimensional
_x = np.linspace(0,10,100)
plt.title('F sen(x)')
plt.xlabel('')
plt.ylabel('oscilación')
#plt.plot(math.sin(math.pi/3))
plt.plot(np.sin(_x),'o')
plt.grid()
#plt.suptitle('$$f(sin)$$', fontsize=16)
```

<b>Fx(Cos(x))</b>

```python
_x = np.linspace(-5, 5, 100)
plt.title('F cos(x)')
plt.xlabel('')
plt.ylabel('oscilación')
#plt.plot(math.sin(math.pi/3))
plt.plot(np.cos(_x),'or')
plt.grid()
#plt.suptitle('$$f(sin)$$', fontsize=16)
```

<b>Fx(Tang(x))</b>

```python
_x = np.linspace(0, 10, 100)
plt.title('F tang(x)')
plt.xlabel('')
plt.ylabel('oscilación')
#plt.plot(math.sin(math.pi/3))
plt.plot(np.tan(_x), color='green')
plt.grid()
#plt.suptitle('$$f(sin)$$', fontsize=16)
```
