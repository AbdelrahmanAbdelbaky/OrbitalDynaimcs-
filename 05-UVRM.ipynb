###### Formulation UVRM     
- 1) $v_{f} = v_{o}+-a\cdot t $   


- 2) $d = v_{o}\cdot t +- \frac{a \cdot t^2}{2} $    


- 3) $v_{f}^2 = v_{o}^2 +- 2\cdot a \cdot d$  


- 4) $d = \frac{v_{o} + v_{f}}{2}\cdot t$  



- 5) $a = \frac {v_{o} + v_{f}}{t} $        

 
- 6) $x = v_{o} +- \frac{1}{2}a (2n-1)$     

#### Exercise 1
An airplane triples its speed after having flown 24m in 6 sec. Find its acceleration.

<b>Res:</b>: 

###### Airplane : 
- $v_{o} = x\frac {m}{\sec}$   

- $v_{f}= 3 \cdot x\frac {m}{\sec}$   

- $d = 24 m $   

- $t = 6\sec $   

-$a = ?$

###### Finding the value of the velocities - Formula 4):  
- $d = \frac{v_{o} + v_{f}}{2}\cdot t$ 

- $24 = \frac{x  +  3x}{2}\cdot 6$          
- $\frac{24\cdot 2}{6} = x + 3x$   
- $8 = 4x$
- $x = \frac{8}{4} = 2\frac{m}{s}$
- $v_{o} = 2\frac {m}{\sec}$    
- $v_{f} = 6\frac {m}{\sec}$   

###### Calculating acceleration - - Formula 5):
- $v_{f} = v_{o}+-a\cdot t $   
- $6 = 2 + a \cdot 6 $   
- $6 -2 = a \cdot 6 $    
- $a = \frac{4}{6} = 0.67\frac{m}{\sec^2}$
